```python
import os
```

```python
rootPath = "D:\Baihm\EISNN\Archive_New"

def delete_ehd_files(directory):
    deleted_count = 0
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith('.ehd'):
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    print(f"已删除: {file_path}")
                    deleted_count += 1
                except Exception as e:
                    print(f"删除失败: {file_path}，原因: {e}")
    if deleted_count == 0:
        print("没有找到 .ehd 文件。")
    else:
        print(f"总共删除了 {deleted_count} 个 .ehd 文件。")

# 示例调用
delete_ehd_files(rootPath)
```
